```python
import numpy as np
from sklearn.datasets import make_regression, make_classification
import matplotlib.pyplot as plt
```

```python
def gradient_descent(learning_rate=0.1, num_iterations=100):
    # Initialize x with a random value
    x = 5  # Starting point, can be any value
    for i in range(num_iterations):
        # Calculate the derivative of the function f(x) = x^2, which is 2*x
        gradient = 2 * x
        # Update the value of x
        x = x - learning_rate * gradient
        # Print the value of x at each iteration
        print(f"Iteration {i+1}: x = {x}, f(x) = {x**2}")
    return x

# Running the gradient descent function
result = gradient_descent(learning_rate=0.1, num_iterations=20)
print(f"Minimum value of x found: {result}")
```

```python
import matplotlib.pyplot as plt

def gradient_descent_plot(learning_rate=0.1, num_iterations=20):
    # Initialize x with a random value
    x = 5  # Starting point, can be any value
    x_values = []
    y_values = []
    
    for i in range(num_iterations):
        # Calculate the function value f(x) = x^2
        y = x ** 2
        x_values.append(x)
        y_values.append(y)
        
        # Calculate the derivative of the function f(x) = x^2, which is 2*x
        gradient = 2 * x
        # Update the value of x
        x = x - learning_rate * gradient
    
    # Plotting the graph
    plt.figure(figsize=(10, 6))
    plt.plot(range(num_iterations), y_values, marker='o')
    plt.title("Gradient Descent Optimization for f(x) = x^2")
    plt.xlabel("Iteration")
    plt.ylabel("f(x)")
    plt.grid(True)
    plt.show()

# Run the plot function
gradient_descent_plot(learning_rate=0.1, num_iterations=20)
```

```python
import numpy as np

def gradient_descent_linear_regression(X, y, learning_rate=0.01, num_iterations=1000):
    m = len(y)  # Number of training examples
    q0, q1 = 0.0, 0.0  # Initial values for intercept and slope
    cost_history = []  # To keep track of the cost at each iteration

    for i in range(num_iterations):
        # Calculate the predictions
        predictions = q0 + q1 * X

        # Calculate the cost function (MSE)
        cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
        cost_history.append(cost)

        # Calculate the gradients
        q0_gradient = (1 / m) * np.sum(predictions - y)
        q1_gradient = (1 / m) * np.sum((predictions - y) * X)

        # Update q0 and q1
        q0 -= learning_rate * q0_gradient
        q1 -= learning_rate * q1_gradient

        # Optionally, print progress
        if i % 100 == 0:
            print(f"Iteration {i}: Cost = {cost}, q0 = {q0}, q1 = {q1}")

    return q0, q1, cost_history

# Sample data (X: independent variable, y: dependent variable)
X = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 2, 4, 5])

# Run gradient descent
q0, q1, cost_history = gradient_descent_linear_regression(X, y, learning_rate=0.01, num_iterations=1000)

print(f"Final values: q0 = {q0}, q1 = {q1}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the gradient descent function with plotting capability
def gradient_descent_linear_regression_with_plot(X, y, learning_rate=0.01, num_iterations=1000):
    m = len(y)  # Number of training examples
    q0, q1 = 0.0, 0.0  # Initial values for intercept and slope
    cost_history = []  # To keep track of the cost at each iteration

    for i in range(num_iterations):
        # Calculate the predictions
        predictions = q0 + q1 * X

        # Calculate the cost function (MSE)
        cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
        cost_history.append(cost)

        # Calculate the gradients
        q0_gradient = (1 / m) * np.sum(predictions - y)
        q1_gradient = (1 / m) * np.sum((predictions - y) * X)

        # Update q0 and q1
        q0 -= learning_rate * q0_gradient
        q1 -= learning_rate * q1_gradient

    return q0, q1, cost_history

# Sample data (X: independent variable, y: dependent variable)
X = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 2, 4, 5])

# Run gradient descent
q0, q1, cost_history = gradient_descent_linear_regression_with_plot(X, y, learning_rate=0.01, num_iterations=1000)

# Plot the cost function
plt.figure(figsize=(10, 6))
plt.plot(cost_history, marker='o')
plt.title("Cost Function Convergence - Gradient Descent")
plt.xlabel("Iteration")
plt.ylabel("Cost (MSE)")
plt.grid(True)
plt.show()

# Plotting the best fit line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, q0 + q1 * X, color='red', label='Best Fit Line')
plt.title("Best Fit Line using Gradient Descent")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()


```
