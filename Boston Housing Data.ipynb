```python
# Iterate over each column in the DataFrame
for column in df2.columns:
    # Find duplicate values in the current column
    duplicates = df2[df2.duplicated(subset=[column])]
    
    # Check if there are any duplicates in the current column
    if not duplicates.empty:
        print(f"Duplicates in column '{column}':")
        print(duplicates[column])
        print()
```

```python
## LOOKING FOR SUM OF DUPLICATES VALUE IN EACH COLUMN

for column in df2.columns:
    dup_sum= df2[df2.duplicated(subset=[column])].value_counts()
    
    print(f"the sum of duplicated value in   {column} is {dup_sum}")
```
